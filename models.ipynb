```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import numpy as np
from numpy.random import seed

import matplotlib.pyplot as plt
from keras.optimizers import Adam
from models import create_phi, create_rho, generate_data

seed(10)
```

```python
n_sets = 70
n_members = 3
member_dim = 5

set_dims = (n_members, member_dim)
member_dims = (member_dim,1)

phi = create_phi(member_dims)
rho = create_rho(member_dims, phi, n_members)
adam = Adam(lr=0.01, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)

rho.compile(optimizer=adam, loss='mean_squared_error', metrics=['accuracy'])
```

```python
train_x, train_y = generate_data(n_sets, set_dims)
val_x, val_y = generate_data(n_sets, set_dims)
test_x, test_y = generate_data(n_sets, set_dims)
```

```python
rho.fit(x=train_x, y=train_y,  epochs=100)
```

```python
rho.evaluate(test_x, test_y)
```
